```python
### 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. 

laliga = pa.read_csv('Laliga.csv', header = 1)

laliga.replace('-', 0, inplace=True)

laliga.tail()
```

```python
### 2. Print all the teams which have started playing between 1930-1980. (5 points)

##Split Debut year into columns
DebutYear = laliga.Debut.str.split(pat="-", n=1, expand=True)

laliga['DebutStartYear'] = DebutYear[0]
laliga['DebutEndYear'] = DebutYear[1]

##check the condition using start year
laliga[(laliga['DebutStartYear'] >= '1930') & (laliga['DebutStartYear'] <= '1980')].Team
```

```python
### 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
### Using the same function, find the team which has maximum and minimum goal difference. (5 points)

# Function to calcualte and include an additional column Goal Difference in the dataframe

def Goal_diff_count ():
        laliga.GoalsFor = laliga.GoalsFor.astype ('int')
        laliga.GoalsAgainst = laliga.GoalsAgainst.astype ('int')
        laliga ['GoalDiff'] = laliga.GoalsFor - laliga.GoalsAgainst
        return laliga[['Team', 'GoalDiff']] 
           

#call the function
Goal_diff_count()

#print team and correspoding goal differences, max/min goal differences
print (laliga[['Team', 'GoalDiff']])
print ('Team with Max Goal Difference - ', laliga.loc[laliga['GoalDiff'].idxmax()].Team)
print ('Team with Min Goal Difference - ', laliga.loc[laliga['GoalDiff'].idxmin()].Team)

```

```python
###Create a new column with name “Winning Percent” and append it to the data set (5 points). 
###Percentage of Winning = (GamesWon / GamesPlayed)*100
###If there are any numerical error, replace it with 0%
###Print the top 5 teams which has the highest Winning percentage

#Convert Data Type
laliga.GamesWon = laliga.GamesWon.astype ('int')
laliga.GamesPlayed = laliga.GamesPlayed.astype ('int')

#calculate and add new column
laliga ['PercentOfWinning'] = ( laliga.GamesWon / laliga.GamesPlayed ) * 100
                
laliga['PercentOfWinning'].fillna(0, inplace=True)

#print (laliga[['Team', 'PercentOfWinning']])

laliga.nlargest(5, 'PercentOfWinning') ['Team']
```
