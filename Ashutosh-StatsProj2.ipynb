```python
import numpy as np
import pandas as pd
```

```python
salesoutput = pd.read_csv ('StatsProj2-SalesOutputData.csv')
salesoutput
```

```python
salesoutput.dtypes
salesoutput.drop(salesoutput.index[-1], inplace=True)
```

```python
print ('Old Scheme Mean', salesoutput.OldScheme.mean())
print ('New Scheme Mean', salesoutput.NewScheme.mean())
```

```python
###2.	Use the five percent significance test over the data to determine the p value to check new scheme has 
### significantly raised outputs? (10 points)
salesoutput
```

```python
# H0 : New Scheme has no effect on output (mean is same as old scheme)
# H1 : New Scheme has an effect on output (mean is greater than old scheme)
from scipy.stats import ttest_1samp
from scipy.stats import ttest_rel

# Since this is for the same set of sales people with old scheme and new scheme, there is a dependency between the two scenrios. 
# so we can either use 1 sample test using the difference in means should be zero 
# or we can use a two-sided test for the null hypothesis that 2 related or repeated samples have identical average (expected) values.


stat1, p1 = ttest_1samp(salesoutput.NewScheme- salesoutput.OldScheme, 0)
stat2, p2 = ttest_rel(salesoutput.OldScheme, salesoutput.NewScheme)

# Since we need one sided (greater than) probability, we need to take half of the probability 

print('Statistics1=%.3f, p1=%.3f' % (stat1, p1/2))
print('Statistics2=%.3f, p2=%.3f' % (stat2, p2/2))

alpha = 0.05
if (p1/2) > alpha:
    print('Same distributions (fail to reject H0). New Scheme does not have significantly raised output')
else:
    print('Different distributions (reject H0).New Scheme has significantly raised output ')
```

# Answer

#To test the Null Hypothesis, we have used ttest on two outcomes and tried validating if the distribution is similar or not. 
#The probability is higher compared to significance level of 5%.

```python
##  4. Suppose it has been calculated that in order for Titan to break even, 
## the average output must increase by £5000 in the scheme compared to the old scheme. 
## If this figure is alternative hypothesis, what is:

##4.1  The probability of a type 1 error? (2.5 points)
```

# 4.1 Answer
#Probability of type 1 error is same as its significance level, i.e. 
#probibility of rejecting the null hypothesis (H0), when it is in fact true. 
#Since this is one sided validation, i.e. only greater than is important, we need to take only half of this signiifcance value.
#Probability of type 1 error = 5/2 = 2.5%

```python
## 4.2 Whzat is the p- value of the hypothesis test if we test for a difference of $5000? (10 points)

# We need to validate the following Hypothesis:
# H0 : New Scheme has no effect on output (diffference in mean is less than or equal to 5 K)
# H1 : New Scheme has an effect on output (diffference in mean is greater than old scheme by 5 K)

# If difference in mean value should be 5000 (or 5 as the values are in 1000s), we should use 1samp test and check for 
# probability for difference in mean as 5

stat1, p1 = ttest_1samp(salesoutput.NewScheme-salesoutput.OldScheme, 5)

# since the value has to be one sided, we need to take half of p value
p=p1/2
print('Statistics1=%.3f, p1=%.3f' % (stat1, p))
```

```python
## 4.3 Power of the test (5 points)
# Calculating Power of Test
# Compute the difference in Means between 2 sample means and divide by pooled Standard Deviation 
# number of Observations/tuples
# Set the alpha value to 0.05 and alternative values 'two-sided' , 'larger' , 'smaller'

 
meandiff  = np.mean(salesoutput.NewScheme) -np.mean(salesoutput.OldScheme)
sp = np.sqrt((29*np.var(salesoutput.OldScheme)+29*np.var(salesoutput.NewScheme)) / 58)
effectsize = meandiff/sp
print (effectsize)
```
